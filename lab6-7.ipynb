```python
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import six
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
%matplotlib inline
```

## Get the Data



```python
df = pd.read_csv("spambase.data")
df.head()
```

```python
df.shape
```

```python
df = df.fillna(df.mean())
```

```python
X = df.iloc[:,:54]
y = df['Decision']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape, X_test.shape
```

##  DT



```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(max_depth = 7, criterion = 'gini')
```

```python
dtree.fit(X_train,y_train)
```

## Predictions



```python
predictions = dtree.predict(X_test)
```

```python
print('Training set score: {:.2f}'.format(dtree.score(X_test,y_test)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#forest = RandomForestClassifier(criterion='gini', n_estimators=100) 
# n_estimators - число деревьев
forest = RandomForestClassifier()
```

```python
forest.fit(X_train, y_train)
```

```python
predictions_rf= forest.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions_rf))
```

```python
print(classification_report(y_test,predictions_rf))
```

## Gridsearch RandFor

```python
#param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 

param_grid = {'criterion' : ['gini', 'entropy'], 'n_estimators':[1, 10, 100, 500], 'max_depth':[5,7,8,9,10], 'max_features':[3,4,5]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
